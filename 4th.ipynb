# Heading missing  data:Replace function

```python
import pandas as pd
import numpy as np
df=pd.read_excel("missing_data.xlsx")
df
```

```python
#-99999 means data is meaning
```

```python
#if you want to replace -99999 with NaN value.
new_df=df.replace(-99999,np.NaN)
new_df
```

```python
weather_data={
    'day':['1\1\2017','1\2\2017','1\3\2017','1\4\2017','1\5\2017','1\6\2017','1\6\2017'],
    'temperature':[32,-99999,28,-99999,32,31,34],
    'windspeed':[6,7,-99999,7,-88888,2,5],
    'event':['rain','sunny','snow','no event','rain','sunny','no event']
}
df=pd.DataFrame(weather_data)
df
```

```python
#if you have two different value and you  want to replace both -99999 and -88888 with NaN
new_df=df.replace([-99999,-88888],np.NaN)
new_df
```

```python
weather_data={
    'day':['1\1\2017','1\2\2017','1\3\2017','1\4\2017','1\5\2017','1\6\2017','1\6\2017'],
    'temperature':[32,-99999,28,-99999,32,31,34],
    'windspeed':[6,7,-99999,7,-88888,2,5],
    'event':['rain','sunny','snow',0,'rain','sunny',0]
}
df=pd.DataFrame(weather_data)
df
```

```python
new_df=df.replace({
    'temperature':-99999,
    'windspeed':[-99999,-88888],
    'event':0
},np.NaN)
new_df
```
